```python
df['Price'] = df['Price'].str.replace(',','')
df['Tax'] = df['Tax'].str.replace(',','')
df
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Tax'] = pd.to_numeric(df['Tax'], errors='coerce')
df.info()
```

```python
df
```

```python
df = df.drop(columns=['Nearest Landmark','Distance to Landmark'],axis = 1)
df = df.dropna(subset=['Rating','Rating Description','Star Rating','Tax'])
df.sort_values('Price', inplace=True) #There is no relation between Tax & Price hence Tax cannot be interpolated & has to be dropped
df['Total Price']=df['Tax']+df['Price']
df
```
